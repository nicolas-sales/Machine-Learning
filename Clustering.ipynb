```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

# Importation du jeu de données make_blobs et création du modèle Kmeans

```python
X, y = make_blobs(n_samples=300, centers=5, n_features=2, random_state=42)

k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
y_pred = kmeans.fit_predict(X)
```

## Visualisation des observations et de leurs labels

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

## Visualisation des observations et des prédictions

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred)

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=20, label='Centres')

plt.legend()
plt.title('KMeans Clustering')
plt.show()
```

# K-moyennes par mini-lots

Divise par 3 à 4 le temps de calcul et permet de partionner les jeux de données volumnieux

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
minibatch_kmeans = MiniBatchKMeans(n_clusters=5, random_state=42)  # Par défaut batch_size=100
minibatch_kmeans.fit(X)
```

## Comparaison de l'inertie entre l'algorithme KMeans et MiniBatchKMeans 

```python
inertias = []
inertias2 = []

k_values = range(1, 10) 

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

for k in k_values:
    minibatch_kmeans = MiniBatchKMeans(n_clusters=k, random_state=42)
    minibatch_kmeans.fit(X)
    inertias2.append(minibatch_kmeans.inertia_)

plt.figure(figsize=(6, 4))
plt.plot(k_values, inertias, color='blue')
plt.plot(k_values, inertias2, color='red')
plt.xlabel('K')
plt.ylabel('Inertie')
plt.title('Variation de l\'inertie en fonction de K')
plt.grid(True)
plt.show()
```

IL y a peu d'écart entre les deux inerties

On observe un coude (K=4) à partir duquel l'inertie ne décroit plus. 4 clusters serait donc un bon choix.

```python
import time

train_times_kmeans = []
train_times_minibatch = []

k_values = range(1, 10)

# Calcul des temps d'entraînement pour KMeans
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # Mesure du temps d'entraînement
    start_time = time.time()
    kmeans.fit(X)
    end_time = time.time()
    
    # Calcul de la durée
    duration = end_time - start_time
    train_times_kmeans.append(duration)

# Calcul des temps d'entraînement pour MiniBatchKMeans
for k in k_values:
    minibatch_kmeans = MiniBatchKMeans(n_clusters=k, random_state=42)
    
    # Mesure du temps d'entraînement
    start_time = time.time()
    minibatch_kmeans.fit(X)
    end_time = time.time()
    
    # Calcul de la durée
    duration = end_time - start_time
    train_times_minibatch.append(duration)

# Affichage des résultats
plt.figure(figsize=(6, 4))
plt.plot(k_values, train_times_kmeans, marker='o', linestyle='-', color='blue', label='KMeans')
plt.plot(k_values, train_times_minibatch, marker='o', linestyle='-', color='red', label='MiniBatchKMeans')
plt.xlabel('K')
plt.ylabel('Durée d\'entraînement (secondes)')
plt.title('Variation de la durée d\'entraînement en fonction de K')
plt.legend()
plt.grid(True)
plt.show()
```

## Coefficient de silhouette

(a-b)/max(a,b)

a : distance moyenne intra cluster

b : moyenne des distances des observations au cluster le plus proche

[-1,1] : 1-> observation à l'intérieur de son propore cluster et loin des autres, 0 -> se situe près d'une frontière entre cluster, -1 -> affectée au mauvais cluster

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, kmeans.labels_)
```

```python
score = []

k_values = range(2, 10) 

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    score.append(silhouette_score(X, kmeans.labels_))

plt.figure(figsize=(6, 4))
plt.plot(k_values, score, color='blue')
plt.xlabel('K')
plt.ylabel('score')
plt.title('score de silhouette en fonction de K')
plt.grid(True)
plt.show()
```

Le graphique confirme qu'un cluster de 4 est un très bon choix

```python
from sklearn.metrics import silhouette_samples

kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

silhouette_avg = silhouette_score(X, y_kmeans)
print(f"Silhouette score moyen : {silhouette_avg}")

# Calcul des échantillons de silhouette pour chaque point
sample_silhouette_values = silhouette_samples(X, y_kmeans)

fig, ax = plt.subplots(figsize=(6, 4))

y_lower = 10
for i in range(4):  
    cluster_silhouette_values = sample_silhouette_values[y_kmeans == i]
    cluster_silhouette_values.sort()
    size_cluster = cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster
    ax.barh(range(y_lower, y_upper), cluster_silhouette_values, height=1.0, edgecolor='none')
    ax.text(-1.05, (y_lower + y_upper) / 2, f"Cluster {i+1}", va='center', ha='center', color='black', fontweight='bold')
    y_lower = y_upper + 10
    
ax.axvline(x=silhouette_avg, color="red", linestyle="--")
ax.set_xlabel("Score de silhouette")
ax.set_ylabel("Cluster")
ax.set_title("Diagramme de Silhouette pour chaque échantillon")
plt.show()
```

La majorité des points sont supérieurs au score de silhouette moyen (pointillés rouges) et ont des scores proches de 1 ce qui signifie qu'ils sont bien regroupés et éloignés des autres clusters

# Clustering pour segmentation d'image

```python
import PIL
from PIL import Image
```

```python
filepath = 'ladybug.jpeg'

image = np.asarray(PIL.Image.open(filepath))
```

```python
image.shape
```

```python
plt.imshow(image)
```

```python
X = image.reshape(-1,3) # -1 : produit des dimensions  -> matrice 2D
kmeans = KMeans(n_clusters=8, random_state=42).fit(X)  # segementation de l'image en 8 groupes de couleurs
segmented_img = kmeans.cluster_centers_[kmeans.labels_]  # Pour chaque pixel, on remplace sa couleur par le centre de son cluster
# Les centres de clusters sont des flottants, il faut les convertir en entiers entre 0 et 255
segmented_img = np.clip(segmented_img, 0, 255).astype(np.uint8)
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img)
```

Les couleurs de l'image ont été converties dans les couleurs propres à leur cluster

# Clustering et apprentissage semi-supervisé

```python
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
```

```python
X_digits, y_digits = load_digits(return_X_y=True)
X_train, y_train = X_digits[:1400], y_digits[:1400]
X_test, y_test = X_digits[1400:], y_digits[1400:]
```

```python
# Sur 50 observations étiquetées
n_labeled = 50
log_reg = LogisticRegression(max_iter=10_000)
log_reg.fit(X_train[:n_labeled], y_train[:n_labeled])
```

```python
log_reg.score(X_test, y_test)
```

```python
log_reg.score(X_train, y_train)
```

```python
k = 50
kmeans = KMeans(n_clusters=k, random_state=42)
X_digits_dist=kmeans.fit_transform(X_train)
representative_digit_idx = np.argmin(X_digits_dist, axis=0)
X_representative_digits = X_train[representative_digit_idx]
```

```python
X_representative_digits.shape
```

```python
fig, axes = plt.subplots(5, 10, figsize=(10, 5))  
axes = axes.ravel()  # Aplati la liste des axes

for i, ax in enumerate(axes):  # Boucle pour chaque axe et image
    ax.imshow(X_representative_digits[i].reshape(8, 8), cmap='gray')  # Affiche l'image
    ax.axis('off')  # Désactive les axes

plt.tight_layout()
plt.show()
```

```python
y_representative_digits = np.array([8, 0, 1, 3, 6, 7 ,5, 4, 1, 8, 2, 3, 9, 5, 3, 9, 1, 7, 4, 1, 4, 6, 9, 7, 5, 2, 2, 1, 3, 3, 6, 0, 4, 9, 1, 1, 7, 4, 2, 4, 1, 3, 9, 7, 8, 9, 6, 5,6,4])  
```

```python
y_representative_digits.shape
```

```python
log_reg = LogisticRegression(max_iter=10_000)
log_reg.fit(X_representative_digits, y_representative_digits)
log_reg.score(X_test,y_test)
```

Normalement, avec l'étiquetage manuelle des images, le score du modèle doit être plus élevé. L'étiquetage peut être un peu faussé vu la qualité des images représentées.

## Propagation des étiquettes

```python
y_train_propagated = np.empty(len(X_train), dtype=np.int64)
for i in range(k):
    y_train_propagated[kmeans.labels_ == i] = y_representative_digits[i]
```

```python
log_reg = LogisticRegression(max_iter=10_000)
log_reg.fit(X_train, y_train_propagated)
log_reg.score(X_test,y_test)
```

## Ignorance des 1% d'observations le plus éloignées de leur centroides

```python
percentile_closest = 99

X_cluster_dist = X_digits_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(k):
    in_cluster = (kmeans.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    cutoff_distance = np.percentile(cluster_dist, percentile_closest)
    above_cutoff = (X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff] = -1

partially_propagated = (X_cluster_dist !=1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
log_reg = LogisticRegression(max_iter=10_000)
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
log_reg.score(X_test,y_test)
```

Le score devrait être légèrement meilleur

```python
(y_train_partially_propagated == y_train[partially_propagated]).mean()
```

Le score avec les étiquettes propagées sont très bonnes
